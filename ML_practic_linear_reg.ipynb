```python
#установим библиотеки

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Загружаем датасет

data_path = '/content/drive/MyDrive/Colab Notebooks/MLdatasets/Salary_dataset.csv'

salary_df = pd.read_csv(data_path) #считываем данные

salary_df #смотрим данные
```

```python
salary_df.describe()
```

```python
# ПОстроим график цен и года эксплуатации
plt.scatter(salary_df.YearsExperience, salary_df.Salary)
```

```python
#Создадим модель линейной регрессии
model = LinearRegression()
```

```python
#Создадим переменные X и y для обучения модели
X = pd.DataFrame(salary_df.YearsExperience)
y = pd.DataFrame(salary_df.Salary)
```

```python
#Обучим модель
model.fit(X,y)
```

```python
#Посмотрим, какой коэффициент у модели и как она выглядит на графике
model.coef_
```

```python
plt.scatter(salary_df.YearsExperience, salary_df.Salary)
plt.plot(X,model.predict(X),color = 'red', linewidth = 1)
```

```python
# Попытка создать недообученную модель

low_model = LinearRegression()
```

```python
low_X = pd.DataFrame(salary_df.YearsExperience[:10])
low_y = pd.DataFrame(salary_df.Salary[:10])

low_model.fit(low_X,low_y)
```

```python
low_model.coef_
```

```python
plt.scatter(salary_df.YearsExperience, salary_df.Salary)
plt.plot(X,model.predict(X),color = 'red', linewidth = 2)
plt.plot(X,low_model.predict(X), color = 'green', linewidth = 2)
```
